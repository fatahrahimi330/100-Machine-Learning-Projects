### Recognizing Handwritten Digits with Multi-Layer Perceptron in Sklearn.

##### 1. Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### 2. Importing Dataset

```python
from sklearn.datasets import load_digits
df = load_digits()
X = df.data
y = df.target
```

##### 3. Display or Plot the images

```python
def display_digits(index):
    nplots = 16
    plt.figure(figsize=(6, 6))
    for j in range(nplots):
        plt.subplot(4, 4, j+1)
        plt.imshow(df.images[index + j], cmap='binary')
        plt.title(df.target[index +j])
        plt.axis('off')
    plt.show()

display_digits(0)
```

##### 4. Train and Test Split

```python
X_train = X[:1000]
X_test = X[1000:]

y_train = y[:1000]
y_test = y[1000:]
```

##### 5. Build and Fit the Model

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(
    hidden_layer_sizes=(15, ),
    activation='logistic',
    solver='sgd',
    alpha=1e-4,
    tol=1e-4,
    random_state=1,
    learning_rate_init=0.1,
    verbose=True
)

model.fit(X_train, y_train)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(model.loss_curve_, 'o-')
plt.title('Loss and Iteration Curve')
plt.xlabel('Number of iteration')
plt.ylabel('Loss')
plt.show()
```

##### 6. Model Evaluation

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4}")
```
