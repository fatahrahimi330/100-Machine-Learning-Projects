### Cartooning an Image using OpenCV
Turning a normal Photo to a funny one.

##### 1. Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

##### 2. Importing the Data

```python
!wget https://raw.githubusercontent.com/fatahrahimi330/100-Machine-Learning-Projects/master/8-Cartooning%20Image/image.png
```

```python
image = cv2.imread('image.png')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

##### 3. Main Part

```python
# Image Preprocessing (Gray and blure)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blure_gray_image = cv2.medianBlur(gray_image, 5)

# Edge
edge = cv2.adaptiveThreshold(blure_gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

# Smooth Color
color = cv2.bilateralFilter(image, 9, 250, 250)

# Combine
cartoon = cv2.bitwise_and(color, color, mask=edge)
```

##### 4. Display the Image

```python
# Note: cv2.imshow() won't work in Jupyter notebooks
# cv2.imshow("Cartoon", cartoon)
cv2.imwrite("cartoon_output.png", cartoon)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
plt.imshow(cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
