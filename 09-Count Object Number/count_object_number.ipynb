### Count Number of Oblects using OpenCV

##### 1. Importing Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

##### 2. Implementation

```python
!wget https://raw.githubusercontent.com/fatahrahimi330/100-Machine-Learning-Projects/master/9-Count%20Object%20Number/image.png -O image.jpg
```

```python
image = cv2.imread('image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
blure = cv2.GaussianBlur(gray, (21, 21), 0)
plt.imshow(blure, cmap='gray')
plt.axis('off')
plt.show()
```

```python
canny = cv2.Canny(blure, 30, 150, 3)
plt.imshow(canny, cmap='gray')
plt.axis('off')
plt.show()
```

```python
dilated = cv2.dilate(canny, (1, 1))
plt.imshow(dilated, cmap='gray')
plt.axis('off')
plt.show()
```

```python

(cnt, hierarchy) = cv2.findContours(
    dilated.copy(),
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_NONE
)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)

plt.imshow(rgb)
plt.axis('off')
plt.show()
```
