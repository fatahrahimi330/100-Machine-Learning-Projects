### Text Detection and Extraction using OpenCV

##### 1. Import Libraries

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

##### 2. Importing Image

```python
!wget https://raw.githubusercontent.com/fatahrahimi330/100-Machine-Learning-Projects/master/11-Text%20Detection%20and%20Extraction/image.png -O image1.png
```

```python
path = "image1.png"
image = cv2.imread(path)
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

##### 3. converting to gray

```python
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
plt.axis('off')
plt.show()
```

##### 4. Extracting Text from Image

```python
extracted_text = pytesseract.image_to_string(img_gray)
print(extracted_text)
```

##### 5. Drawing Boundry around detected texts

```python
data = pytesseract.image_to_data(img_rgb, output_type=pytesseract.Output.DICT)
num_boxes = len(data['level'])
for i in range(num_boxes):
    (x, y, w, h) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h),(255, 0, 0), 2)
```

##### 6. Display Image with Bounding Box

```python
plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.title("Image with Text Bounding Boxes")
plt.axis("off")
plt.show()
```
