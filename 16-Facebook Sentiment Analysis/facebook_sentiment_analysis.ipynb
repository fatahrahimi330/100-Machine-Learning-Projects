### Facebook Sentiment Analysis using NLTK in Python

##### 1. Importing Libraries

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import io
import unicodedata
import numpy as np
import re
import string
from numpy import linalg
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import webtext
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
# Download required NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('vader_lexicon')
```

```python
# Loading the Data
with open('kindle.txt', encoding='UTF-8') as f:
    text = f.read()

# Tokenization
print("=" * 50)
print("Word Tokens:")
print("=" * 50)
word_tokens = word_tokenize(text)
print(word_tokens[:20])  # Print first 20 tokens

print("\n" + "=" * 50)
print("Sentence Tokens:")
print("=" * 50)
sent_tokens = sent_tokenize(text)
print(sent_tokens[:5])  # Print first 5 sentences

# Stemming
print("\n" + "=" * 50)
print("Stemming (first 10 words):")
print("=" * 50)
porter_stemmer = PorterStemmer()
for i, w in enumerate(word_tokens[:10]):
    print("Actual: {} Stem: {}".format(w, porter_stemmer.stem(w)))

# Lemmatization
print("\n" + "=" * 50)
print("Lemmatization (first 10 words):")
print("=" * 50)
wordnet_lemmatizer = WordNetLemmatizer()
for i, w in enumerate(word_tokens[:10]):
    print("Actual: {} Lemma: {}".format(w, wordnet_lemmatizer.lemmatize(w)))

# POS Tagging
print("\n" + "=" * 50)
print("POS Tagging (first 20 tokens):")
print("=" * 50)
pos_tags = nltk.pos_tag(word_tokens[:20])
print(pos_tags)

# Sentiment Analysis
print("\n" + "=" * 50)
print("Sentiment Analysis:")
print("=" * 50)
sid = SentimentIntensityAnalyzer()

with open('kindle.txt', encoding='UTF-8') as f:
    for line in f.read().split('\n'):
        if line.strip():  # Only process non-empty lines
            print("\nText: {}".format(line))
            scores = sid.polarity_scores(line)
            for key in sorted(scores):
                print('{}: {}'.format(key, scores[key]), end=' ')
            print()
```
